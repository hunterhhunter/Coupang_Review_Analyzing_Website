```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/morphed_data.csv', encoding='cp949', index_col=0)
data
```

```python
good = data[data.iloc[:, 1]==1]
bad = data[data.iloc[:, 1]==0]
```

```python
bad
```

```python
sampled_good = good.sample(n=len(bad)*3, replace=False)

print(len(sampled_good))
print(sampled_good)
```

```python
all_data = pd.concat([sampled_good, bad], axis=0)
all_data
```

```python
all_data = all_data.sample(frac=1).reset_index(drop=True)
all_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(all_data.iloc[:, 0], all_data.iloc[:, 1], stratify=all_data.iloc[:, 1])
```

```python
train_data = pd.concat([X_train, y_train], axis=1).reset_index(drop=True)
test_data = pd.concat([X_test, y_test], axis=1).reset_index(drop=True)

train_data.to_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/train.csv', encoding='cp949', index=0)
test_data.to_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/test.csv', encoding='cp949', index=0)
```
