```python
import pandas as pd
from collections import Counter
from collections import defaultdict
```

```python
data = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/data/전처리_완_데이터/review_여성패션.csv', encoding='cp949', index_col=0)
data
```

```python
data = data.dropna(subset='리뷰')
```

```python
from tqdm import tqdm_notebook
word_list = []

for sentence in tqdm_notebook(data['리뷰']):
    word_list += sentence.split(' ')
word_list
```

```python
count = Counter(word_list)
result = count.most_common(1000)
```

```python
print(f'count: {result}')
```

```python
words_1000 = []
cnt = 0
for i in count.most_common(1000):
    words_1000.append(i[0])
words_1000
```
