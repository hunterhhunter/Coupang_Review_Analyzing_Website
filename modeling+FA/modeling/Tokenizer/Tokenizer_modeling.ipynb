```python
full_data = pd.concat([data1, data2, data3, data4], axis=0, ignore_index=True)
full_data
```

```python
def labeling(score):
    if score < 3:
        return 0
    elif score == 5:
        return 1
```

```python
label_data = full_data[(full_data['구매자 평점'] > 4) | (full_data['구매자 평점'] < 3)]
label_data
```

```python
label_data['긍/부정'] = label_data['구매자 평점'].apply(labeling)
label_data
```

```python
# Data Tokenizing

from tensorflow.keras.preprocessing.text import Tokenizer
input_series = label_data['리뷰'].astype(str)
```

```python
tokenizer = Tokenizer(num_words=None, char_level=False, lower=False)
tokenizer.fit_on_texts(list(input_series))
word_index = tokenizer.word_index

print(f'전체에서 {len(word_index)}개의 고유한 토큰을 찾았습니다.')
print(f'word_index: {word_index}')
print(f'vocab_size: {len(word_index)}')
```
