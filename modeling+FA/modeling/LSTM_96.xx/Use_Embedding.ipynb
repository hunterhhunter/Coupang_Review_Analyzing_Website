```python
train_data = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/train.csv', encoding='cp949')
test_data = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/test.csv', encoding='cp949')
test_data

final_test = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/morphed_data(청소기).csv', encoding='cp949', index_col=0)
```

```python
final_test
```

```python
maxlen = train_data.iloc[:, 0].apply(lambda x: len(x.split())).max()
print(f'maxlen: {maxlen}')
```

```python
maxlen = test_data.iloc[:, 0].apply(lambda x: len(x.split())).max()
print(f'maxlen: {maxlen}')
```

```python
maxlen = final_test.iloc[:, 0].apply(lambda x: len(x.split())).max()
print(f'maxlen: {maxlen}')
```

```python
tokened_train = tokenizer.texts_to_sequences(list(train_data.iloc[:, 0]))
tokened_test = tokenizer.texts_to_sequences(list(test_data.iloc[:, 0]))
tokened_final = tokenizer.texts_to_sequences(list(final_test.iloc[:, 0]))
tokened_train
```

```python
# Padding
from tensorflow.keras.preprocessing.sequence import pad_sequences

train_pad = pad_sequences(tokened_train, maxlen=maxlen, padding='post')
test_pad = pad_sequences(tokened_test, maxlen=maxlen, padding='post')
final_pad = pad_sequences(tokened_final, maxlen=maxlen, padding='post')
```

```python
import numpy as np
train_label = np.array(train_data.iloc[:, 1])
test_label = np.array(test_data.iloc[:, 1])
final_label = np.array(final_test.iloc[:, 1])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Embedding, Flatten, LSTM, Bidirectional

embedding_dim = 100
max_words = 29836
model = Sequential()

model.add(Embedding(input_dim=max_words, output_dim=embedding_dim))
model.add(Bidirectional(LSTM(64, return_sequences=True, dropout=0.3)))
model.add(Bidirectional(LSTM(16, dropout=0.3)))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
from keras.callbacks import ModelCheckpoint

modelpath = 'C:/Users/slugg/Documents/GitHub/three_idiot/model/models/LSTM_checkpoint/'+"{epoch:d}-{val_loss:.4f}.h5"

checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_acc', verbose=1)
```

```python
history = model.fit(train_pad, train_label, batch_size=32, epochs=10, validation_data=[test_pad, test_label], callbacks=[checkpointer])
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
print('Train accuracy of each epoch:', np.round(acc, 3))
print('Validation accuracy of each epoch:', np.round(val_acc, 3))
epochs = range(1, len(val_acc) + 1)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import rhinoMorph
rn = rhinoMorph.startRhino()

maxlen = 1085

def predict_one_review(sentence: str):
    morphed_data = rhinoMorph.onlyMorph_list(rn, sentence, pos=['NNG', 'NNP', 'VV', 
                                                        'VA', 'XR', 'IC', 'MM', 'MAG', 'MAJ'], eomi=True)
    morphed_data = ' '.join(morphed_data)
    
    tokened_data = tokenizer.texts_to_sequences(morphed_data)
    paded_data = pad_sequences(tokened_data, maxlen=maxlen, padding='post')

    result = model.predict(paded_data)
    
    return result, [morphed_data, tokened_data]
```
