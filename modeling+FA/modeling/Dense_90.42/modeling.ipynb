```python
import pickle
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
data = pd.read_csv('C:/Users/slugg/Documents/GitHub/three_idiot/전처리_데이터/morphed_data.csv', index_col=0, encoding='cp949')
data
```

```python
input_texts = data.iloc[:, 0]
```

```python
tokenizer_path = 'C:/Users/slugg/Documents/GitHub/three_idiot/model/Tokenizer/third_tokenizer.pickle'

with open(tokenizer_path, 'rb') as file:
    tokenizer = pickle.load(file)
```

```python
word_index = tokenizer.word_index

print('\n전체에서 %s개의 고유한 토큰을 찾았습니다.' % len(word_index))
print('word_index: ', word_index)
print('vocab_size: ', len(word_index))
```

```python
encoder_input = tokenizer.texts_to_sequences(list(input_texts))

print('인코딩 결과')
print(input_texts[0], encoder_input[0])
```
