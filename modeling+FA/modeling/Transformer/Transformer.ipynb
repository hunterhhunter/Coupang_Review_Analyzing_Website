```python
"""마스킹 (1/3) 패딩"""
# seq 값이 padding 0일 때만 1을 출력하고 그 외에는 0을 출력
# 마스킹 대상을 1.0으로 만들어 이후 -10억을 곱해 softmax함수를 통해 역전
# 입력 (batch_size, seq_len) -> return (batch_size, 1, 1, seq_len)
# 이후 attention의 (batch_size, heads, encoder_len, seq_len)에 합산됨

def create_padding_mask(seq):
    mask = tf.cast(tf.math.equal(seq, 0), tf.float32)
    return mask[: ,tf.newaxis, tf.newaxis, :]
```

```python
"""마스킹 (2/3) 순방향"""

def create_look_ahead_mask(size):
    mask = 1 - tf.linalg.band_part(tf.ones((size, size)), -1, 0) # 하삼각행렬 만들기 (tf.linalg.band_part(2D데이터, 0, 1) -> 상삼각행렬)
    return mask
```

```python
"""마스킹 (3/3) 결합"""
def create_masks(inp):
    enc_padding_mask = create_padding_mask(inp)
    return enc_padding_mask
```

```python
"""포지셔널 인코딩 (1/2)"""
def get_angles(pos, i, d_model):
    # PE = sin(pos/(10000^(2i/d_model))) 수식의 sin안의 수식
    # pos 는 포지션에 데한 인덱스 위치 리스트, i는 차원 배열, d_model은 단어의 차원 수
    angle_rate = 1 / np.power(10000, (2 * i/2) / np.float32(d_model))
    return pos * angle_rate
```

```python
"""포지셔널 인코딩 (2/2)"""
def positional_encoding(position, d_model):
    #앞에서 만든 get_angles()를 이용해 단어의 위치 정보를 생성
    angle_rads = get_angles(np.arange(position)[:, np.newaxis],
                            np.arange(d_model)[np.newaxis, :], 
                            d_model)
    # 단어 표현 공간이 짝수(2i)인 경우 sin, 홀수(2i+1)인 경우 cos함수를 사용하여 구분
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    angle_rads[:, 1::2] = np.cos(angle_rads[:, 1::2])
    pos_encoding = angle_rads[np.newaxis, ...]

    return tf.cast(pos_encoding, dtype=(tf.float32))
```

```python
"""스케일 내적 어텐션"""
def scaled_dot_product_attention(q, k, v, mask):
    matmul_qk = tf.matmul(q, k, transpose_b=True) # Q행렬과 K행렬을 내적연산해 Attention Score를 구한다 (Transpose_b는 k를 전치)

    dk = tf.cast(tf.shape(k)[-1], tf.float32) # k행렬의 차원 수(열의 수)를 구한다
    scaled_attention_logits = matmul_qk / tf.math.sqrt(dk) # Key 벡터의 차원 수의 제곱근으로 나눠 크기를 줄인다 (Scaling)

    if mask is not None:
        scaled_attention_logits += (mask * -1e9) 
        # 매우 큰 음수를 mask와 곱한뒤 더해 softmax를 거치며 매우 작은 값은 0으로 마스킹되어
        # 자신보다 뒤에 나오는 단어는 참조되지 못한다. 그 외의 양의 값은 확률 정보가 된다.(하삼각) 순방향 마스크 어텐션이 수행된 것
    attention_weights = tf.nn.softmax(scaled_attention_logits, axis=-1)

    # 확률값 Attention Score에 Value 벡터로 가중합을 수행
    output = tf.matmul(attention_weights, v)
    
    return output, attention_weights
```
