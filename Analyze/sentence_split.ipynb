```python
marks = ["?", "!", "。", "‘", "’", "“", "”", "`", "\'",
        "\"", "(", ")", "{", "}", "[", "]", "─", "『", "』", 
        ",", "ㆍ", "·", "ᆞ", ":", ";", "/", "…", "_", "~", 
        "∼", "∽", "□", "■", "▶", "◀", "◆", "▲", "◇", "◈", 
        "☎", "【", "】", "+", "-", "=", "±", "÷", "×", "*", 
        "^", ">", "<", "｜", "|", "％", "%", "&", "￦", "₩", 
        "\\", "\t", "\r\n", "\n", "＄", "$", "¥", "￥", "£", 
        "￡", "°", "㎞", "㎏", "@", "©", "ⓒ", "↑", "|", "#", 
        "♥", "♡", "★", "☆", "♪", "♬", 'ㅋ', 'ㅠ', 'ㅜ']
```

```python
def remove_mark(sentence):
    sentence_no_mark = sentence
    for mark in marks:
        sentence_no_mark = sentence_no_mark.replace(mark, '')
    return sentence_no_mark
```

```python
data['리뷰'] = data['리뷰'].astype(str).apply(remove_mark)
```

```python
from multiprocessing import Pool
import numpy as np

def process_chunk(chunk):
    result_list = []
    for index, row in chunk.iterrows():
        product = row['제품명']
        review = row['리뷰']
        splited_sentence_list = kss.split_sentences(review)
        result_list.extend([{'상품명': product, '리뷰': sentence} for sentence in splited_sentence_list])
    return result_list

def parallelize_dataframe(df, func, num_partitions=4):
    # 데이터프레임을 여러 청크로 나누기
    df_split = np.array_split(df, num_partitions)
    
    # 프로세스 풀 생성
    with Pool(num_partitions) as pool:
        # 각 프로세스에 함수를 적용하고 결과를 모음
        result_list = pool.map(func, df_split)
    
    # 모든 결과를 하나로 합치기
    return pd.DataFrame([item for sublist in result_list for item in sublist])

# 중간 결과를 출력하는 함수
def print_partial_results(result_list):
    total_sentences = sum(len(item) for item in result_list)
    print(f"{total_sentences} 문장 처리 완료")

# 원본 데이터프레임을 여러 청크로 나누어 처리
new_dataframe = parallelize_dataframe(dd, process_chunk)
print_partial_results(new_dataframe)

# 새로운 데이터프레임 확인
print(new_dataframe.shape)
```
